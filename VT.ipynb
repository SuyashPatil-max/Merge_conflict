```python
import torch
import torchvision
from torchvision import transforms
from PIL import Image
from pathlib import Path

# =========================
# PATHS (EDIT THESE)
# =========================
INPUT_DIR = Path("/content/WhatsApp Video 2026-01-31 at 12_35_19 AM_mp4")     # folder with input images
OUTPUT_DIR = Path("/content/cnn_output")      # folder to save cropped images
OUTPUT_DIR.mkdir(exist_ok=True)

# =========================
# DEVICE
# =========================
device = "cuda" if torch.cuda.is_available() else "cpu"

# =========================
# LOAD PRETRAINED CNN
# =========================
model = torchvision.models.detection.fasterrcnn_resnet50_fpn(
    weights="DEFAULT"
)
model.to(device)
model.eval()

# =========================
# TRANSFORM
# =========================
transform = transforms.ToTensor()

# =========================
# PROCESS EACH IMAGE
# =========================
for image_path in sorted(INPUT_DIR.iterdir()):

    if image_path.suffix.lower() not in [".jpg", ".jpeg", ".png"]:
        continue

    try:
        # -------------------------
        # LOAD IMAGE
        # -------------------------
        image = Image.open(image_path).convert("RGB")
        image_tensor = transform(image).unsqueeze(0).to(device)

        # -------------------------
        # INFERENCE
        # -------------------------
        with torch.no_grad():
            output = model(image_tensor)[0]

        boxes = output["boxes"]
        scores = output["scores"]

        if len(boxes) == 0:
            print(f"‚ö†Ô∏è No object detected in {image_path.name}")
            continue

        # -------------------------
        # PICK BEST OBJECT
        # -------------------------
        best_idx = torch.argmax(scores)
        x1, y1, x2, y2 = boxes[best_idx].int().tolist()

        # -------------------------
        # CROP OBJECT
        # -------------------------
        cropped = image.crop((x1, y1, x2, y2))

        # -------------------------
        # SAVE RESULT
        # -------------------------
        output_path = OUTPUT_DIR / image_path.name
        cropped.save(output_path)

        print(f"‚úÖ Saved: {output_path.name}")

    except Exception as e:
        print(f"‚ùå Failed on {image_path.name}: {e}")

print("üéâ Batch processing completed!")
```
