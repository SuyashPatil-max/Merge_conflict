```python
import torch
import torchvision
from torchvision import transforms
from PIL import Image
from pathlib import Path


INPUT_DIR = Path("/content/WhatsApp Video 2026-01-31 at 12_35_19 AM_mp4")
OUTPUT_DIR = Path("/content/cnn_output")
OUTPUT_DIR.mkdir(exist_ok=True)

device = "cuda" if torch.cuda.is_available() else "cpu"

model = torchvision.models.detection.fasterrcnn_resnet50_fpn(
    weights="DEFAULT"
)
model.to(device)
model.eval()

transform = transforms.ToTensor()

for image_path in sorted(INPUT_DIR.iterdir()):

    if image_path.suffix.lower() not in [".jpg", ".jpeg", ".png"]:
        continue

    try:
        image = Image.open(image_path).convert("RGB")
        image_tensor = transform(image).unsqueeze(0).to(device)

        with torch.no_grad():
            output = model(image_tensor)[0]

        boxes = output["boxes"]
        scores = output["scores"]

        if len(boxes) == 0:
            print(f" No object detected in {image_path.name}")
            continue
        best_idx = torch.argmax(scores)
        x1, y1, x2, y2 = boxes[best_idx].int().tolist()

        cropped = image.crop((x1, y1, x2, y2))
        output_path = OUTPUT_DIR / image_path.name
        cropped.save(output_path)

        print(f" Saved: {output_path.name}")

    except Exception as e:
        print(f" Failed on {image_path.name}: {e}")

print(" Batch processing completed!")
```
